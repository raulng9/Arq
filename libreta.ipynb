```python
import csv
import folium
from folium.plugins import MarkerCluster
import pandas as pd
from pyproj import Transformer

#Coordinates in CSV are in ETRS89 - UTM zone 30N, we need to convert them to standard GPS ones (4326)
transformer = Transformer.from_crs("25830", "EPSG:4326")
initial_coord = [43.533, -6.583]

with open('places.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    my_map = folium.Map(location = initial_coord, zoom_start = 13)
    for row in csv_reader:
        if line_count == 0:
            line_count += 1
        else:
            row[1] = row[1].replace(',','.')
            row[2] = row[2].replace(',','.')
            if line_count == 1:
                initial_coord = list(transformer.transform(row[1], row[2]))
            folium.Marker(list(transformer.transform(row[1], row[2])), popup = row[0]).add_to(my_map)
            line_count += 1
    print(f'processed {line_count} items')


#print(transformer.transform(231779.108, 4820035.35))

#Define coordinates of where we want to center our map
boulder_coords = [40.015, -105.2705]



#Create the map
#my_map = folium.Map(location = boulder_coords, zoom_start = 13)

#Display the map
my_map
```
